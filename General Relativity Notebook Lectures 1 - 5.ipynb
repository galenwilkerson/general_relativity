# Lecture 1 - Equivalence principle

## Equivalence principle:

**Gravity is in some sense the same as acceleration**


> General relativity is a theory not just about gravity, but about geometry.

```python
Image(filename = "images/elevator.png", width=200, height=200)
```

In an elevator.

2 coordinate frames:

1. Earth's coordinate frame  ($z$)
2. inside the elevator  ($z'$)

> Note, here our velocities are 'slow', in that we (ironically) ignore relativistic effects of speed of light, etc.  (Here we are just building up basic ideas of reference frames.)

$z$ = height of a point from earth

$z'$ = height of a point from the floor of the elevator

$L(t)$ is the height of the elevator from the earth.

Both $z$ and $z'$ are **inertial reference frames**, which means they are related by uniform velocity.  

This means we can say $L(t) = vt$, the distance of the elevator to the ground at time $t$ is just the velocity * time.

#### Coordinate transformation 
(from global reference frame to elevator reference frame)

* $z' = z - L(t) \implies z' = z - vt$
> Since $z' + L(t) = z$

* $t' = t$
* $x' - x$

Now let's look at Newton's Law of Motion

$F = m\ddot{z}$   is the equation of motion in the global frame ($z$)
>(same as $F = ma$, where $\ddot{z}$ is the second derivative of $z$ with respect to time, i.e. acceleration)

What is the equation of motion in the the elevator frame ($z'$)?

Just have to calculate the acceleration in the 'prime' frame.

$\ddot{z}' = \ddot{z}$

> (Since 
> $$
\frac{d^2}{dt^2} z - vt = \frac{d}{dt} \dot{z} - v = \ddot{z}
$$ 

That is, the acceleration in the two frames of reference are the same, obtaining

$$
F = m\ddot{z}'
$$

According to Newton, the laws of physics are the same in all inertial frames.

### Now, let's look at accelerated reference frame

$L(t)$ will be increasing in an accelerated way. 

Let
$L(t) = \frac 12 gt^2$  
> (This $\frac12$ coefficient is arbitrary, chosen to make the derivatives cleaner.)

Taking derivatives with respect to time, we get

$\dot L = gt$

$ \ddot L = g$

For our **coordinate transformation**, we get:

* $z' = z - L(t) = z - \frac 12 gt^2$
* $t' = t$
* $x' - x$

For our equation of motion, we get

> $\ddot{z}' = \ddot{z} - g$

> (Since $\dot z = \dot z - gt$)

$\implies g + \ddot z' = \ddot z$

$\implies F = m\ddot z = m(g + \ddot z') = m\ddot z' + mg$

$\implies F = m\ddot z' + mg$

We then shift $mg$ to the other side of the equation to make it look like a Newton equation, mass x acceleration is equal to something.  
That 'something' is the **force in the primed (elevator) frame of reference.**

$-mg + F = m \ddot z'$

This primed force has a 'fictitious' term which is just the mass times the acceleration of the elevator.

This looks like a uniform gravitational field.

**The special feature of gravity is that gravitational forces are proportional to mass.**, which has deep implications.

The motion of an object in a gravitational field does not depend on its mass.  E.g. motion of Earth about the sun does not depend on its mass.
Before Einstein, this was considered a coincidence.  Einstein showed that they are indistinguishable.

Given a metric tensor, how do you decide if there is a transformation which will make it look like the kronecker delta ($\delta$)?

$$
\delta_{ij} = \begin{cases}
0 &\text{if } i \neq j,   \\

$$

That is, can we write

$$
dS^2 = \sum_{m,n} g_{m,n}(x) dx^m dx^n

as

$$
\sum_{m,n}\delta_{m,n} dx^m dx^n \text{   ?}

>if you think of the kronecker delta as a matrix (technically it is not), it would look like the identity matrix:

\begin{bmatrix}

Suppose there are two sets of coordinates:

* $x^m(y)$ - that is, $x$s are vector-valued functions of the vectors $y$
* $y^n(x)$ - and also, $y$s are vector-values functions of the vectors $x$

and a small separation $dx^m$ between two points.

What's the corresponding separation of the $y$ coordinate?

$$
dy^m = \sum_p \frac{\partial y^m}{\partial x^p}dx^p
$$

This is the transformation of a tensor.

```python
Image(filename = "images/coord_change.png", width=300, height=300)
```

$dx$ and $dy$ represent a little vector located at some point in the space.

We would like to be able to convert from one set of coordinates to another.

If you change coordinates, you change the components of a vector, even if you are not changing the vector.

### Contravariant vector

$$
V'^m = \sum_p \frac{\partial y^m}{\partial x^p} V^p
$$

**Basically, we have a vector in one coordinate frame and want to transform it to another coordinate frame.  
We take a bunch of partial derivatives of one coordinate frame to get the other one.**

> **Every quantity which under a transformation of coordinates, transforms like the coordinate differentials is called a contravariant tensor.**

> source: https://math.stackexchange.com/questions/8170/intuitive-way-to-understand-covariance-and-contravariance-in-tensor-algebra

#### Einstein notation (basically a vectorized notation)

We can write the above without the summation.  Basically we know we have to do the inner product since the $p$ dimensions match.

$$
V'^m = \frac{\partial y^m}{\partial x^p} V^p
$$

### Covariant vector

Suppose we have a scalar function $S$  
> (what we think of as a normal everyday function, not a vector-valued function)

The gradient of a scalar is a vector, but doesn't transform in the same way.

If we know the gradient

$$
\frac{\partial S(x)}{\partial x^p}
$$


$$
\frac{\partial S}{\partial y^m} = \sum_p \frac{\partial S}{\partial x^p} \frac{\partial x^p}{\partial y^m} \space{        }\text{   (The chain rule)}
$$  

and again, in Einstein's "vectorized" notation, don't write the summation $\sum_p$.
$$
\frac{\partial S}{\partial y^m} = \frac{\partial S}{\partial x^p} \frac{\partial x^p}{\partial y^m} 
$$  

If there is a repeated index on the right hand side (in this case, $p$), that does not appear on the left-hand side, it must be a summation.

Calling the gradient of the scalar function $W$, we get:

$$
W_m' = \frac{\partial x^p}{\partial y^m} W_p
$$

### Summary:

**Contravariant vector**:

$$
V'^m = \frac{\partial y^m}{\partial x^p} V^p
$$

**Covariant vector**:

$$
W_m' = \frac{\partial x^p}{\partial y^m} W_p
$$

**Contravariant transformations are used for change of coordinates within the surface.**

> That is, we want to write a vector along the surface in terms of some new coordinates.

> Characteristic of velocities.

**Covariant transformations are used for change of coordinates of gradients tangent to the surface.**

> That is, we have a scalar function (like some surface) and we want to write a gradient in terms of new coordinates.

> Characteristic of gradients.

### Tensors with more than one index

#### Contravariant

** Transformation of a contravariant tensor rank two **

Start with two contravariant vector and take their product:

$$
V^mU^n = T^{mn}
$$

In 3-space, $V^m$ and $U^n$ each have 3 components, while $T^{mn}$ has 9 components.

>(By Einstein notation rules, on the left-hand side, we see the index $m$ and $n$ only once, so there is no inner product between $V$ and $U$, instead an outer product.)

> Also notice that $T^{mn}$ is not the same as $T^{nm}$.

Since $T^{mn}$ has two indices ($m,n$), it is called a **tensor of rank 2.**

We just compose two contravariant transformations above:

$$
(T')^{mn} = V'^m U'^q = \frac{\partial y^m}{\partial x^p} V^p \frac{\partial y^n}{\partial x^q} U^q
$$

$$
= \frac{\partial y^m}{\partial x^p} \frac{\partial y^n}{\partial x^q}  V^p U^q
$$

Therefore,

$$
(T')^{mn} =  \frac{\partial y^m}{\partial x^p} \frac{\partial y^n}{\partial x^q} T^{pq}
$$

**for rank 3 contravariant tensor**:

$$
(T')^{lmn} =  \frac{\partial y^m}{\partial x^p} \frac{\partial y^n}{\partial x^q} \frac{\partial y^l}{\partial x^r} T^{pqr}
$$

#### Covariant

**Transformation of covariant tensor rank two**

Again, composing two covariant transformations above:  

> Note here we use subscripts$_{mn}$ instead of superscripts.

$$
(T')_{mn} = W'_m Z'_n = \frac{\partial x^p}{\partial y^m} W_p  \frac{\partial x^q}{\partial y^n} Z_q
$$

$$
 = \frac{\partial x^p}{\partial y^m}  \frac{\partial x^n}{\partial y^n}  W_p Z_n
$$

Giving us

$$
(T')_{mn} = \frac{\partial x^p}{\partial y^m}  \frac{\partial x^n}{\partial y^n} T_{pn}
$$

\implies  g'(y)_{pq} = g(x)_{mn} \frac{\partial x^m}{\partial y^p} \frac{\partial x^n}{\partial y^q}
$$

This is a tensor with 2 covariant indices.  Check, it is a tensor and transforms as a tensor.

It is symmetric and non-zero eigenvalues

non-zero eigenvalues because a zero-eigenvalue would correspond to a vector of zero length.

Therefore it is invertible.

$$
g^{mn}
